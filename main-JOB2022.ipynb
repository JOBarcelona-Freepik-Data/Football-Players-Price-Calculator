```python
def dummies_position_get (df):
    dummies_position=pd.get_dummies(df.position)
    new_df = pd.concat([df, dummies_position],axis=1)
    new_df.drop('position',axis=1, inplace=True)
    return new_df

def dummies_nation_get (df):
    dummies_nation=pd.get_dummies(df.nation)
    new_df = pd.concat([df, dummies_nation],axis=1)
    new_df.drop('nation',axis=1, inplace=True)
    return new_df
```

```python
columns_feat= list(players_train_dummies_nat_pos.columns)
columns_feat.pop(0)
columns_feat.pop(6)
columns_feat.pop(3)
columns_feat.pop(2)
FEATURES= columns_feat
```

```python
train=players_train_dummies_nat_pos.copy()
train.drop(['player_ID','team','league'], axis = 1, inplace = True)
target=players_train_dummies_nat_pos['price']
TARGET='price'
players_train_dummies_nat_pos.head()
```

```python
N_split = 2
kf = KFold(n_splits=N_split)

def objective(trial):
    params = {
                #'iterations' : 10000, replaced by early stopping
                'eval_metric': 'SMAPE', 
                'use_best_model': True,
                'random_seed' : 1,
                'learning_rate' :trial.suggest_loguniform('learning_rate', 0.01, 0.3),
                "depth": trial.suggest_int("depth", 1, 15),
                'l2_leaf_reg' :trial.suggest_loguniform('l2_leaf_reg', 1e-8, 20),
                'random_strength' : trial.suggest_loguniform('random_strength', 1, 50),
                'grow_policy':trial.suggest_categorical ('grow_policy', ['Lossguide','SymmetricTree']),
                'max_bin': trial.suggest_int("max_bin", 20, 500),
                'min_data_in_leaf':trial.suggest_int('min_data_in_leaf', 1, 100),
                "bootstrap_type": trial.suggest_categorical("bootstrap_type", ["Bayesian", "Bernoulli"])
            }
    
    if params['grow_policy'] == 'Lossguide':
        params['max_leaves']:trial.suggest_int('max_leaves', 1, 100)
    if params["bootstrap_type"] == "Bayesian":
        params["bagging_temperature"] = trial.suggest_float("bagging_temperature", 0, 10)
    elif params["bootstrap_type"] == "Bernoulli":
        params["subsample"] = trial.suggest_float("subsample", 0.1, 1)
        
    
    score_list = []
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(train, train[TARGET])):
        X_tr = train.loc[train_idx][FEATURES]
        X_va = train.iloc[val_idx][FEATURES]
        
        # Preprocess the data
        X_tr_f = X_tr # Use the correct preprocessing if required (normalization...)
        y_tr = train[TARGET].loc[train_idx].values

        X_va_f = X_va # Use the correct preprocessing if required (normalization...)
        y_va = train[TARGET].loc[val_idx].values

        # Train the model
        model = CatBoostRegressor(**params) 
        model.fit(        
                X_tr_f,
                y_tr,
                eval_set =[( X_va_f,y_va)],
                verbose =0,
                early_stopping_rounds = 200)

        # Predictions
        y_va_pred = model.predict(X_va_f)
        score = mean_squared_error(y_va, y_va_pred,squared = True)
        score_list.append(score)
        
    return sum(score_list) / len(score_list)

study4 = optuna.create_study(direction='minimize')
study4.optimize(objective, n_trials=20)
```

```python
def evaluate_Regression_models(model_name, model,x_test,y_test):
    prediction = model.predict(x_test)
    
    MAE=mean_absolute_error(y_test,prediction)
    MSE =  np.sqrt(mean_squared_error(y_test,prediction))
    r2 = r2_score(y_test,prediction)
                                  
    print(f"Mean Absolute Error: {MAE}")
    print(f"Mean Squared Error : {MSE}") 
    print(f"r2 : {r2}")
          
    df=pd.DataFrame({"Model":model_name, "MAE":MAE, "MSE":MSE, "r2":r2})
                                 
    return df

```
